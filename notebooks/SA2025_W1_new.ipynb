### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
# 1. Compute the average mpg for each (model_year, origin) combination
avg_mpg = (
    cars
    .groupby(['model_year', 'origin'])['mpg']
    .mean()
    .reset_index()
)

# 2. Create the line plot
plt.figure(figsize=(10, 6))
sns.lineplot(
    data=avg_mpg,
    x='model_year',
    y='mpg',
    hue='origin',
    marker='o'
)

# 3. Tidy up the axes and legend
plt.title('Evolution of Average MPG Over Years, by Origin')
plt.xlabel('Model Year')
plt.ylabel('Average MPG')
plt.legend(title='Origin')
plt.tight_layout()
plt.show()
```

### Create a Seaborn scatterplot (or PairGrid) where:

X = horsepower

Y = weight

Color by: origin

Size by: mpg

Hue order = ['japan', 'europe', 'usa']

Add meaningful plot titles and axis titles.

```python
# 1. Define the plotting style (optional)
sns.set_style("whitegrid")

# 2. Create the scatterplot
plt.figure(figsize=(10, 6))
scatter = sns.scatterplot(
    data=cars,
    x='horsepower',
    y='weight',
    hue='origin',
    hue_order=['japan', 'europe', 'usa'],
    size='mpg',
    sizes=(40, 200),          # adjust min/max point sizes as needed
    alpha=0.7,                # slight transparency for overlap clarity
    palette='deep'            # or any other Seaborn-compatible palette
)

# 3. Add titles and axis labels
plt.title('Horsepower vs. Weight Colored by Origin and Sized by MPG', fontsize=14)
plt.xlabel('Horsepower (hp)', fontsize=12)
plt.ylabel('Weight (lbs)', fontsize=12)

# 4. Tweak the legend
#   - Place legend outside plot for readability
#   - Separate hue and size legends
hue_legend = scatter.legend_
hue_legend.set_title('origin')
plt.legend(
    title='mpg',
    loc='upper left',
    bbox_to_anchor=(1.02, 1),
    labelspacing=0.5
)

# 5. Show the plot
plt.tight_layout()
plt.show()
```

### We define a consistent car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
# 1. Compute the average mpg for each car name in each year
yearly_mpg = (
    cars
    .groupby(['name', 'model_year'])['mpg']
    .mean()
    .reset_index()
)

# 2. For each car name, compute:
#    - appearances  = number of distinct years (count)
#    - mpg_std      = standard deviation of those peryear average mpgs
#    - avg_mpg_all  = average mpg across all years
stats = (
    yearly_mpg
    .groupby('name')['mpg']
    .agg(
        appearances='count',
        mpg_std='std',
        avg_mpg_all='mean'
    )
    .reset_index()
)

# 3. Filter to cars with at least 2 appearances and std(mpg) < 1.0
consistent = stats[
    (stats['appearances'] >= 2) &
    (stats['mpg_std'] < 1.0)
].copy()

# 4. Identify the lowest variation (minimum mpg_std) among those
min_std = consistent['mpg_std'].min()
lowest_variation = consistent[consistent['mpg_std'] == min_std]

# 5. (Bonus) Sort all consistent cars by:
#       1) appearances (descending)
#       2) avg_mpg_all    (descending)
sorted_consistent = consistent.sort_values(
    by=['appearances', 'avg_mpg_all'],
    ascending=[False, False]
)

# 6. Display results
print("Cars with the lowest mpg variation (std < 1.0):")
print(lowest_variation[['name', 'appearances', 'avg_mpg_all', 'mpg_std']])

print("\n---\nAll consistent cars sorted (by appearances, then avg_mpg):")
print(sorted_consistent[['name', 'appearances', 'avg_mpg_all', 'mpg_std']])
```
